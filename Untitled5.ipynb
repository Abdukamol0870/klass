```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.datasets import make_blobs

# Sun'iy ma'lumot yaratish (2 ta klass)
X, y = make_blobs(n_samples=50, centers=2, random_state=6)
y = np.where(y == 0, -1, 1) # SVM uchun 0 ni -1 ga o'giramiz

# Modelni o'qitish
model = SVC(kernel='linear', C=1.0)
model.fit(X, y)

# Vizualizatsiya
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='bwr', edgecolors='k')
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Qaror chegarasini chizish
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
P = model.decision_function(xy).reshape(XX.shape)

ax.contour(XX, YY, P, colors='k', levels=[-1, 0, 1], alpha=0.5, linestyles=['--', '-', '--'])
plt.title("Sklearn: Linear SVM")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs

# 1. Ma'lumotlarni yaratish (2 ta klass, chiziqli ajraladigan)
X, y = make_blobs(n_samples=50, centers=2, random_state=6)
y = np.where(y == 0, -1, 1)

class PerfectLinearSVM:
    def __init__(self, C=1.0, lr=0.001, iters=2000):
        self.C = C
        self.lr = lr
        self.iters = iters
        self.w = None
        self.b = None

    def fit(self, X, y):
        n_samples, n_features = X.shape
        self.w = np.zeros(n_features)
        self.b = 0

        for _ in range(self.iters):
            for i, x_i in enumerate(X):
                # SVM sharti: y_i * (w*x_i - b) >= 1
                condition = y[i] * (np.dot(x_i, self.w) - self.b) >= 1

                if condition:
                    # To'g'ri: Faqat regulyarizatsiya (yo'lakni kengaytirish)
                    self.w -= self.lr * (2 * 1/self.C * self.w)
                else:
                    # Xato yoki Margin ichida: Yo'lakni to'g'rilash va xatoni jazolash
                    self.w -= self.lr * (2 * 1/self.C * self.w - np.dot(x_i, y[i]))
                    self.b -= self.lr * y[i]

    def get_hyperplane_value(self, x, v):
        # f(x) = w0*x + w1*y - b = v  => y = (v + b - w0*x) / w1
        return (v + self.b - self.w[0] * x) / self.w[1]

# 2. Modelni o'qitish
# C parametrini oshirish marginni (yo'lakni) qat'iyroq qiladi
model = PerfectLinearSVM(C=10.0, iters=2500)
model.fit(X, y)

# 3. Vizualizatsiya
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='bwr', edgecolors='k')

x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
x_axis = np.linspace(x_min, x_max, 100)

# Qaror chegarasi (Decision Boundary) - v=0 bo'lgan holat
plt.plot(x_axis, model.get_hyperplane_value(x_axis, 0), 'k-', label='Qaror chegarasi')

# Ijobiy va Salbiy Marginlar (v=1 va v=-1)
# Sklearn kabi nuqtali chiziqlar bilan chizamiz
plt.plot(x_axis, model.get_hyperplane_value(x_axis, 1), 'k--', alpha=0.5, label='Marginlar (y=±1)')
plt.plot(x_axis, model.get_hyperplane_value(x_axis, -1), 'k--', alpha=0.5)

plt.ylim(X[:, 1].min() - 1, X[:, 1].max() + 1)
plt.title("Manual SVM: Sklearn kabi mukammal chiziqlar")
plt.legend()
plt.grid(alpha=0.2)
plt.show()
```
